接下来介绍下绘制各图案的函数

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
img = np.zeros((512,512,3),np.uint8)
```

解释下上面的代码：创建空白图像。img = np.zeros((512,512,3),np.uint8)。我来解释下各个参数：第一个就是指的是创建一个三维数组，决定了数组的形状。对于第二个参数就是决定里面存储数据的类型

绘制直线函数：cv.line(img,start,end,color,thickness)

```python
cv.line(img,(0,0),(500,500),(255,0,0),5)
plt.imshow(img[:,:,::-1])
```

绘制圆形：cv.circle(img,centerpoint,r,color,thickness),当thickness=-1时就是实心圆

```python
cv.circle(img,(250,250),50,(0,255,0),5)
plt.imshow(img[:,:,::-1])
```

绘制矩形：cv.rectangle(img,leftupper,rightupper,color,thickness)

```python
cv.rectangle(img,(0,0),(200,200),(0,0,255),5)
plt.imshow(img[:,:,::-1])
```

 图像添加文字：cv.putText(img,text,station,font,fontsize,color,thickness,cv.LINE_AA)

```python
cv.putText(img,'test',(50,400),cv.FONT_HERSHEY_SIMPLEX,4,(255,255,255),2,cv.LINE_AA)
plt.imshow(img[:,:,::-1])
```

接下来介绍修改图像中的像素点

```python
img2=np.zeros((500,500,3),np.uint8)
channel=img2[100,100]
print(channel)
```

channel变量是一个长度为3的一维数组，分别对应bgr三个通道，里面对应相应通道的数值

```python
BC=img2[100,100,0]
print(BC)
```

BC就是取的位置为(100,100)像素的blue通道的数值

```python
img2[100,100]=[255,0,0]
plt.imshow(img2[:,:,::-1])
```

修改特定像素的属性，操作等同于数组的相关修改操作

```python
img3=cv.imread('D:\image\preview.jpg',1)
print(img3.shape,img3.size,img3.dtype,sep=' ')
```

以上就是图像的相关属性

```python
b,g,r=cv.split(img3)
img_merge=cv.merge((b,g,r))
plt.imshow(b,'gray')
```

```python
plt.imshow(g,'gray')
```

```python
plt.imshow(r,'gray')
```

```python
plt.imshow(img_merge[:,:,::-1])
```

对于通道的拆分当然可以用数组的方法将各个通道提取出来，也可以用split函数，显然对于代码实现来说更加简洁

接下来介绍色彩空间的转换

```python
img_gray=cv.cvtColor(img3,cv.COLOR_BGR2GRAY)
plt.imshow(img_gray,'gray')
```

彩色转黑白

```python
img_hsv=cv.cvtColor(img3,cv.COLOR_BGR2HSV)
plt.imshow(img_hsv)
```

HSV 是另一种颜色空间，通道顺序是 H（色相）、S（饱和度）、V（明度），而非 BGR 或 RGB 的 “颜色通道” 顺序。
