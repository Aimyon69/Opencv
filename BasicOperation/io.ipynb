接下来我演示下如何打开一个图片

```python
import numpy as np
import cv2 as cv
img=cv.imread('D:\image\preview.jpg',0)
cv.imshow('image',img)
cv.waitKey(5000)
cv.destroyAllWindows()
```

现在对上面的代码进行解析，对于imread函数，就是读取图像的函数，他的两个参数，第一个就是要打开的图片的路径，第二个是读取方式的选取。我现在来解释各个参数对应的方式：
0-以灰度模式加载图像
1-以彩色模式加载图像，任何图像的透明度都将被忽略
-1-包括alpha通道加载图像模式
接下来我演示其他两个参数的效果

```python
import numpy as np
import cv2 as cv
img=cv.imread('D:\image\preview.jpg',1)
cv.imshow('image',img)
cv.waitKey(5000)
cv.destroyAllWindows()
```

```python
import numpy as np
import cv2 as cv
img=cv.imread('D:\image\preview.jpg',-1)
cv.imshow('image',img)
cv.waitKey(5000)
cv.destroyAllWindows()
```

对应的图像效果可以去查看当前目录下的png文件，前缀表示打开模式

对于上面的最后两个函数进行解释：
waitKey函数，就是维持程序当前状态一段时间，里面的参数是以毫秒为单位。如果参数为0的话，就是等待用户的一个键盘输入
destroyAllWindows函数就是关闭所有的窗口，为了防止窗口崩溃无响应

接下来介绍另一种图像显示的方法：

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
img=cv.imread('D:\image\preview.jpg',1)
plt.imshow(img[:,:,::-1])
```

这里对上面的代码进行解释，对于pyplot来说，图像一般是rgb这个顺序构建通道，但是opencv是按bgr顺序构建通道。所以如果直接显示cv读取的图像的话，一定会失真。我们需要将读取到的图像经行通道反转
在上面的代码中[:,:,::-1]，这是python的数组切片语法。前两个:代表选取整列(行)保持不变，::-1对当前维度进行翻转
这样bgr就变成了rgb

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
img=cv.imread('D:\image\preview.jpg',-1)
plt.imshow(img[:,:,::-1])
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
img=cv.imread('D:\image\preview.jpg',0)
plt.imshow(img,'gray')
```

解释下上面的代码:
默认情况下，plt.imshow() 会使用默认的彩色映射（如 viridis），此时即使输入的是灰度图（二维数组），也会被映射成彩色（不同亮度对应不同颜色），导致灰度图显示异常（变成 “伪彩色图”）
加上 cmap='gray' 后，会强制使用灰度映射规则：数值越小（越接近 0）显示为黑色，数值越大（越接近 255）显示为白色，中间值显示为不同深度的灰色，符合人眼对灰度图的认知
“将二维的灰度图像数组，按照‘黑 - 白渐变’的规则映射为像素颜色并显示，确保灰度图看起来是正常的黑白效果”
