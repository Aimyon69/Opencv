形态学操作:

连通性:<span style="color:red;">1.</span>两个像素的位置是否相邻</span><span style="color:green;">(4,8,D[对角]邻接)</span>。<span style="color:red;">2.</span>两个像素的灰度值是否满足特定的相似准则。(具体的定义参考数字图像处理)

混合连通(m连通)：第一种情况：p，q像素是4连通的，那么它俩一定是混合连通的。情况二：p，q对角连通且两像素点的4领域的交集的像素点不符合相似准则

膨胀：就是使图像中高亮的部分扩张，效果图比原图拥有更大的高亮区域，是局部求最大值的操作。具体的操作如下图所示：

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
img_dilate=cv.imread('D:\Code\Opencv\Morphology\dilate.jpg')
plt.imshow(img_dilate[:,:,::-1])
```

介绍cv的操作函数<span style="color:green;">dilate(img,kernel,iterations)</span>:img:要操作的图像。kernel:核结构，上图中的结构B。ite：腐蚀的次数

腐蚀：让原图像中的高亮地区被蚕食，效果图比原图拥有更小的高亮地区，是求局部最小值的操作，原理图如下:
